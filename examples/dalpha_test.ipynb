```python
import nbodypy as npy
import matplotlib.pyplot as plt
import numpy as np
import limepy
from limepy import limepy
```

```python
w0=[2.,4.,6.,8.]
```

```python
cluster=[]
clabel=['W0=2','W0=4','W0=6','W0=8']
rm0=[]
phi0=[2.,4.,6.,8.]
cluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=2.))
cluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=4.))
cluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=6.))
cluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=8.))


for c in cluster:
    c.key_params()

    print(c.rm,c.mtot)
    rm0.append(c.rm)
```

```python
for i in range(0,len(cluster)):
    c=cluster[i]
    plt.plot(c.x,c.y,'o',alpha=0.5,label=clabel[i])

plt.legend()
plt.show()
```

```python
for i in range(0,len(cluster)):
    c=cluster[i]
    kwargs={'label':clabel[i]}

    rprof,pprof,nprof=npy.rho_prof(c)        

    if i==0:
        npy.nlplot(rprof,pprof,log=True,label=clabel[i],xlabel='r \ (pc)',ylabel=r'$\rho \ (Msun/pc^3)$')
    else:
        npy.nlplot(rprof,pprof,log=True,label=clabel[i],overplot=True)
plt.legend()
plt.show()
plt.close()
```

```python
mscluster=[]
mscluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=2.))
mscluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=4.))
mscluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=6.))
mscluster.append(npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=8.))
```

```python
for c in mscluster:
    print(c.m[0],c.m[-1])
    c.m=npy.get_segregated_imf(c,-1.,-1., mlimits = [0.1, 1.0],do_random=True)
    c.key_params()   
    print(c.m[0],c.m[-1])
```

```python
for i in range(0,len(cluster)):
    c=cluster[i]
    mc=mscluster[i]
    kwargs={'label':clabel[i]}

    rprof,pprof,nprof=npy.rho_prof(c)        
    mrprof,mpprof,mnprof=npy.rho_prof(mc)        

    if i==0:
        npy.nlplot(rprof,pprof,log=True,label=clabel[i],xlabel='r \ (pc)',ylabel=r'$\rho \ (Msun/pc^3)$')
    else:
        npy.nlplot(rprof,pprof,log=True,label=clabel[i],overplot=True)
        
    npy.nlplot(mrprof,mpprof,'--',log=True,label='MS',overplot=True)
    
    print(c.rm,c.mtot,mc.rm,mc.mtot)

plt.legend()
plt.show()
plt.close()
```

```python
for c in mscluster:
    c.mass_function(0.1,1.,nmass=20,plot=True)
    plt.show()
    plt.close()

    c.alpha_prof(0.1,1.,nrad=20,nmass=20,plot=True)
    plt.show()
    plt.close()
```

```python
dalpha=[0.,-0.5,-1.,-1.5]
alpha50=[-1.35,-1.,-0.65,-0.3]

da=np.empty((len(dalpha),len(alpha50),len(cluster)))
ag=np.empty((len(dalpha),len(alpha50),len(cluster)))
a50=np.empty((len(dalpha),len(alpha50),len(cluster)))

for i in range(0,len(dalpha)):
    for j in range(0,len(alpha50)):
        for k in range(0,len(phi0)):
            c=npy.setup_cluster('king',N=10000,M=1000,rt=100,phi0=phi0[k])  
            c.key_params()

            plt.close()
            print(dalpha[i],alpha50[j])
            c.m=npy.get_segregated_imf(c,dalpha[i],alpha50[j], mlimits = [0.1, 1.0],do_random=True)
            c.key_params()
            
            rmrat=rm0[k]/c.rm
            c.x*=rmrat
            c.y*=rmrat
            c.z*=rmrat
            c.key_params()

            c.mass_function(0.1,1.,nmass=20,plot=True)
            plt.show()
            plt.close()
            
            c.alpha_prof(0.1,1.,nrad=20,nmass=20,plot=True)
            plt.show()
            plt.close()
            
            da[i,j,k]=c.dalpha
            ag[i,j,k]=c.alpha
            a50[i,j,k]=c.alpha50
            
            
```

```python
plt.figure(figsize=(10,10))
for k in range(0,len(cluster)):
    plt.subplot(2,2,k+1)

    for j in range(0,len(alpha50)):
        
        x=da[:,j,k]
        y=a50[:,j,k]/ag[:,j,k]

        plt.plot(x,y,label=r'$\alpha_{50} = %f$' % alpha50[j])
        
    if k==0:
        plt.legend()

    plt.ylim(0.5,1.5)
    plt.ylabel(r'$\alpha_{50}/\alpha$')
    plt.xlabel(r'$\delta_\alpha$')
    plt.gca().invert_xaxis()
    plt.title(clabel[k])

plt.show()
plt.close()
    
```

```python
plt.figure(figsize=(10,10))
for i in range(0,len(dalpha)):
    plt.subplot(2,2,i+1)
    
    for j in range(0,len(alpha50)):
        
        x=w0
        y=ag[i,j,:]

        plt.plot(x,y,label=r'$\alpha_{50} = %f$' % alpha50[j])
        plt.plot(x,a50[i,j,:],'k--',alpha=0.5)

    if i==0:
        plt.legend()

    #plt.ylim(0,1.1)
    plt.ylabel(r'$\alpha_{global}$')
    plt.xlabel(r'$W_0$')
    plt.title(r'$\delta_\alpha$ = %f' % dalpha[i])

plt.savefig('alpha_test.png')
plt.show()
plt.close()
```
