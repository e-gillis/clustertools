```python
import nbodypy as npy
import matplotlib.pyplot as plt
import numpy as np
```

```python
cluster=npy.setup_cluster('limepy',N=8594,phi0=5.75,g=1,M=4297.,rt=145,project=True)
cluster0=npy.setup_cluster('limepy',N=8594,phi0=5.75,g=1,M=4297.,rt=145,project=True)
cluster.key_params()
cluster0.key_params()

print(cluster.rm,cluster.mtot)
print(cluster0.rm,cluster0.mtot)

cluster.energies()
cluster0.energies()
print(cluster.qvir,cluster0.qvir)
```

```python
cluster.m=npy.get_imf(mlimits = [0.1,1.,100.],alphas=[-1.35,-2.35],ntot=cluster.ntot)
cluster.key_params()
print(cluster.rm,cluster.mtot,cluster0.m[0],cluster.m[0])
```

```python
cluster.mass_function(0.1,1.,nmass=50,plot=True)
plt.show()
plt.close()

cluster.mass_function(1.1,100.,nmass=50,plot=True)
plt.show()
plt.close()

cluster.mass_function(0.1,100.,nmass=50,plot=True)
plt.show()
plt.close()
```

```python
imf=npy.get_imf(mlimits = [0.1,1.,100.],alphas=[-1.35,-2.35],mtot=4297.0)
print(cluster.rm,cluster.mtot)
print(cluster0.rm,cluster0.mtot)
```

```python
imf=npy.get_segregated_imf(cluster,-1.,-1., mlimits = [0.1, 1.0],do_random=True)
cluster.key_params()
print(cluster.rm,cluster.mtot)
print(cluster0.rm,cluster0.mtot)
```

```python
cluster.mass_function(0.1,1.,nmass=50,plot=True)
plt.show()
```

```python
cluster.alpha_prof(0.1,50.,rmin=0.,rmax=100,nrad=20,nmass=50,plot=True)
plt.show()
```

```python
npy.rho_prof(cluster0,plot=True)
kwargs={'overplot':True}
npy.rho_prof(cluster,plot=True,**kwargs)
plt.show()
```

```python
print(cluster0.rm)
print(cluster.rm)
rmrat=cluster0.rm/cluster.rm
```

```python
cluster.x*=rmrat
cluster.y*=rmrat
cluster.z*=rmrat
cluster.key_params()
print(cluster.rm)
```

```python
npy.rho_prof(cluster0,plot=True)
kwargs={'overplot':True}
npy.rho_prof(cluster,plot=True,**kwargs)
plt.show()
```

```python
cluster.energies()
print(cluster.qvir)
```
